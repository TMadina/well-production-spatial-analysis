```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('wells.csv')
x=df['depth_m']
y=df['production']
plt.scatter(x,y)
z=np.polyfit(x,y,1)
p=np.poly1d(z)
plt.plot(x,p(x))
plt.xlabel('Depth(m)')
plt.ylabel('Production')
plt.title('Depth vs Production with trend line')
plt.show()
```

```python
correlation=df['depth_m'].corr(df['production'])
print('Correlation:',correlation)
```

```python
plt.scatter(x,y)
plt.plot(x,p(x))
plt.xlabel('Depth(m)')
plt.ylabel('Production')
plt.title(F'Depth vs Production(corr={correlation:.2f})')
```
